```python
import numpy as np
import matplotlib.pyplot as plt

RESULT_PATH = "result.txt"
MATRIX1_PATH = "matrix1.txt"
MATRIX2_PATH = "matrix2.txt"
STATS_PATH = "stats.txt"

def verify_matrices(matrix1, matrix2, expected_result):
    """   ."""
    return np.array_equal(np.dot(matrix1, matrix2), expected_result)

def check():
    try:
        matrix1 = np.loadtxt(MATRIX1_PATH, int, skiprows=1)
        matrix2 = np.loadtxt(MATRIX2_PATH, int, skiprows=1)
        res_matrix = np.loadtxt(RESULT_PATH, int, skiprows=1)

        if verify_matrices(matrix1, matrix2, res_matrix):
            print("Matrices are equal")
        else:
            print("Matrices are not equal")
    except IOError as e:
        print(f" /: {e}")
    except ValueError as e:
        print(f" : {e}")

def showStats():
    try:
        with open(STATS_PATH) as f:
            data = [line.split() for line in f]
        x, y = zip(*((int(size), float(time)) for size, time in data))

        fig, ax = plt.subplots()
        ax.set_title("   ( )")
        ax.plot(x, y, "go-")
        ax.set_xlabel(" ")
        ax.set_ylabel("")
        for i, j in zip(x, y):
            ax.text(i, j, str(j), ha='center', va='bottom')
        plt.show()
    except IOError as e:
        print(f" /: {e}")
    except ValueError as e:
        print(f" : {e}")

if __name__ == "__main__":
    check()
    showStats()
```
